```python
fig = go.Figure( )

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[10] _27MAB10CP006XQ10'],
    name="Давление пара в камере нижнего отопительного отбора (ПСГ №1) т.1 ,кгс/см2"
))


fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[60] _27NDD10CL001_OPC'],
    name="L конд. в ПСГ №1 ,мм",
    yaxis="y2"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[61] _27NDD10CL004_OPC'],
    name="L конд. в сборн. конд. ПСГ №1 ,мм",
    yaxis="y2"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[68] _27LCJ66CF001.1'],
    name="F конд. за КЭН №1,2 ПСГ №1 ,т/ч",
    yaxis="y5"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[70] _26LCJ66CZ071_XQ01.1'],
    name="УП ИМ РК на сливе конд. из ПСГ №1 за Н1 Н2 (КР2) ,%",
    yaxis="y3"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[50] _27LCA13CP001.1'],
    name="P конд. на нап. КЭН N2",
    yaxis="y4"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[46] _27LCJ63CP001.1'],
    name="P конд. на нап. КЭН N2_2",
    yaxis="y4"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[45] _27LCJ62CP001.1'],
    name="P конд. на нап. КЭН №1 ПСГ№1",
    yaxis="y4"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data['[44] _27LCJ64CP001.1'],
    name="P конд. на нап. КЭН НЗ ПНД №1",
    yaxis="y4"
))

# Create axis objects
fig.update_layout(
    xaxis=dict(
        domain=[0.01, 0.99]
    ),
    yaxis=dict(
        title="Давление в отборах",
        # titlefont=dict(
        #     color="#1f77b4"
        # ),
        # tickfont=dict(
        #     color="#1f77b4"
        # )
    ),
    yaxis2=dict(
        title="Уровень",
        # titlefont=dict(
        #     color="#ff7f0e"
        # ),
        # tickfont=dict(
        #     color="#ff7f0e"
        # ),
        anchor="free",
        overlaying="y",
        side="left",
        position=0.0
    ),
    yaxis3=dict(
        title="Положение РК КЭН",
        # titlefont=dict(
        #     color="#d62728"
        # ),
        # tickfont=dict(
        #     color="#d62728"
        # ),
        anchor="x",
        overlaying="y",
        side="right"
    ),
    yaxis4=dict(
        title="Давление на напоре КЭН",
        # titlefont=dict(
        #     color="#9467bd"
        # ),
        # tickfont=dict(
        #     color="#9467bd"
        # ),
        anchor="free",
        overlaying="y",
        side="right",
        position=0.995
    ),
    yaxis5=dict(
        title="Расход конд",
        # titlefont=dict(
        #     color="#9467bd"
        # ),
        # tickfont=dict(
        #     color="#9467bd"
        # ),
        anchor="free",
        overlaying="y",
        side="right",
        position=0.999
    )
)

# Update layout properties
fig.update_layout(
    title_text="Многосевой график",
   width=10000, height=1000
)

fig.show()
```
