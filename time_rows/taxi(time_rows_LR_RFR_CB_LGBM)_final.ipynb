## Обучение

Добавим новых признаков даты, времени через функцию:

```python
def make_feat(data,roll,shift):
    df = data.copy()
    for i in range(1,shift+1):
        df['lag_' + str(i)] = df['num_orders'].shift(i)
    df['rolling'] = df['num_orders'].shift().rolling(roll).mean()
    #df['rolling_sum'] = df['num_orders'].shift().rolling(roll).sum()
    df = df.dropna()
    df['dayofweek'] = df.index.dayofweek
    #df['day'] = df.index.day
    df['month'] = df.index.month
    df['hour'] = df.index.hour
    df['weekofyear'] = df.index.weekofyear
    
    return df
```

```python
cat_features = ['dayofweek','month','day']
```

Разделим датасет на обучающую и тестовую:

```python
data_train,data_test = train_test_split(data,shuffle=False,test_size=0.1)
```

Напишем две функции - одну для подбора наиболее оптимальных ширины окна скользящего среднего и количества лагов, вторую для обучения без подбора этих параметров:

```python
def train_iter(data,roll_max,shift_max,model,model_type):
    best_RMSE = 10000
    for r in range(1,roll_max+1):
        for s in range(1,shift_max+1):
            # Добавляем фичи
            data_m = make_feat(data,r,s)
            # Разбиваем на обучающую, тестовую, фичи и таргет
            features = data_m.drop(['num_orders'],axis='columns')
            target = data_m['num_orders']
            # Разобьем на обучающую и валидационную выборки
            features_train,features_valid,target_train,target_valid = train_test_split(features,target,shuffle=False,test_size=0.2)
            # Учим модель
            if model_type == 'CatBoost':
                model.fit(features_train,target_train,plot=True)
            elif model_type == 'LGMB':
                model.fit(features_train,target_train,categorical_feature=cat_features)
            else:
                model.fit(features_train,target_train)            
            predicts = model.predict(features_valid)
            RMSE = (mean_squared_error(target_valid,predicts)**0.5)
            if RMSE < best_RMSE:
                best_RMSE = RMSE
                best_model = model
                best_roll = r
                best_shift = s
            if s%10 == 0:
                print('Not yet,','roll=',r,'shift=',s,'RMSE=',RMSE)
    print('RMSE valid на алгоритме',model_type,'=',best_RMSE)
    return best_model,best_roll,best_shift
```

// Вышеуказанная функция была применена для подбора оптимальных параметров для модели линейной регрессии. Но, поскольку подбор осуществляется достаточно долго, далее данная функция не задествована.

```python
def train(data,roll_max,shift_max,model,model_type):
    # Добавляем фичи
    data_m = make_feat(data,roll_max,shift_max)
    # Разбиваем на обучающую, тестовую, фичи и таргет
    features = data_m.drop(['num_orders'],axis='columns')
    target = data_m['num_orders']
    # Разобьем на обучающую и валидационную выборки
    features_train,features_valid,target_train,target_valid = train_test_split(features,target,shuffle=False,test_size=0.1)
    # Учим модель
    if model_type == 'LGMB':
        model.fit(features_train,target_train)#,categorical_feature=cat_features)
    else:
        model.fit(features_train,target_train)            
    predicts = model.predict(features_valid)
    RMSE = (mean_squared_error(target_valid,predicts)**0.5)
    print('RMSE valid на алгоритме',model_type,'=',RMSE)
    del(features_train,features_valid,target_train,target_valid)
    return model,roll_max,shift_max
```

Напишем функцию для проверки моделей на тестовой выборке:

```python
def testing(data,model,roll,shift):
    data = make_feat(data,roll,shift)
    features = data.drop(['num_orders'],axis='columns')
    target = data['num_orders']
    predicts = model.predict(features)
    RMSE = (mean_squared_error(target,predicts)**0.5)
    print('RMSE на тестовой выборке:',RMSE)
    print('roll=',roll)
    print('shift=',shift)
    return RMSE
```

Обучим модель линейной регрессии:

```python
model_LR,roll_LR,shift_LR = train(data_train,100,96,LinearRegression(normalize=False,n_jobs=-1),'линейная регрессия')
```

Обучим модель случайного леса:

```python
model_RF,roll_RF,shift_RF = train(data_train,24,8,RandomForestRegressor(n_jobs=-1,n_estimators=100,max_depth=20),'случайный лес')
```

Обучим модель CatBoost:

```python
model_CB,roll_CB,shift_CB = train(data_train,48,96,CatBoostRegressor(loss_function='RMSE',verbose=10),'CatBoost')#,cat_features=cat_features
```

Обучим модель LGBM

```python
model_LGBM,roll_LGBM,shift_LGBM = train(data_train,100,96,LGBMRegressor(learning_rate=0.25,max_depth=20),'LGBM')
```

Обучены модели линейной регрессии, случайного леса, CatBoost, LGBM.

## Тестирование

Проведем тестирование обученных моделей на тестовой выборке с тем же преобразованием данных:

```python
RMSE_LR = testing(data_test,model_LR,roll_LR,shift_LR)
```

```python
RMSE_RF = testing(data_test,model_RF,roll_RF,shift_RF)
```

```python
RMSE_CB = testing(data_test,model_CB,roll_CB,shift_CB)
```

```python
RMSE_LGBM = testing(data_test,model_LGBM,roll_LGBM,shift_LGBM)
```

Соберем все результаты воедино:

```python
results = pd.DataFrame({'Название алгоритма':['Linear Regression','Random Forest Regressor','CatBoost','Ligth Gradient Boost Machine'],'RMSE':[RMSE_LR,RMSE_RF,RMSE_CB,RMSE_LGBM]})
results
```
